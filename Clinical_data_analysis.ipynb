## Data Analysis

### 1. Patient Distribution by Condition

This analysis shows the number of patients affected by each medical condition.

```python
df.groupby("Condition")["Patient_ID"].count()
```

- Patient distribution across conditions is nearly uniform, with multiple conditions having the same highest count. 
This indicates no single condition dominates the dataset.

### 2. Patient Distribution by Gender
This analysis shows the distribution of patients by gender.

```python
df.groupby("Gender")["Age"].count()
```

- Female patients are slightly higher compared to female patients, indicating a marginally higher hospital admission rate among females.

### 3. Average Treatment Cost by Condition

This analysis helps identify which medical conditions have higher treatment costs.

```python
df.groupby("Condition")["Cost"].mean().sort_values(ascending = False)
```

- Cancer has the highest average treatment cost, followed by prostate cancer. 
This indicates that these conditions contribute to higher financial burden on patients and healthcare systems.

### 4. Average Length of Stay by Condition
This analysis shows ehich condtions require longer hospitalization.

```python
df.groupby("Condition")["Length_of_Stay"].mean().sort_values(ascending = False)
```

- Cancer patients have the highest average length of stay, indicating longer hospitalization requirements and higher resource utilization, helping hospitals plan bed occupancy.

### 5. Patient Satisfaction Analysis
This analysis shows how satisfied patients are with their treatment and hospital services.

```python
df.groupby("Satisfaction")["Condition"].count()
```

- Most patients fall under satisfaction level 4, indicating generally positive patient experience across conditions, with minimal variation between conditions.

### 6. Patient Outcome Analysis
This analysis shows treatment effectiveness based on patient outcomes.

```python
df.groupby("Outcome")["Patient_ID"].count()
```

- A higher number of patients have recovered compared to other outcomes,indicating overall treatment effectiveness.

## Final Summary

This project analyzed hospital patient data to understand treatment costs, patient distribution, hospital stay duration, satisfaction levels, and outcomes.
